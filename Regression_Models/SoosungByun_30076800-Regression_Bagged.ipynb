### **Data Curation**

```python
import pandas as pd

df = pd.read_csv("/content/Volumetric_features.csv")
df.head()
```

```python
df = df.drop(columns=["S.No", "dataset"])
df = df.drop_duplicates()
df = df.dropna()
df.head()
```

```python
from sklearn.model_selection import train_test_split
y = df["Age"]
x = df.loc[:,df.columns!="Age"]
```

```python
# Standardizing

# mean_std_x=(x-x.mean())/x.std()
# mean_std_x.head()
```

```python
# Normalizing
# min_max_x=(x-x.min())/(x.max()-x.min())
# min_max_x.head()
```

```python
# Split training and test data
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=142)
```

### **Bagged Regression**

```python
# Importing necessary library
from sklearn.ensemble import BaggingRegressor
from sklearn.linear_model import LinearRegression

# Defining Base Estimator
BEst = LinearRegression()

# Bagged Regressor
BR = BaggingRegressor(estimator=BEst, n_estimators=50, random_state=42)

# Training
BR.fit(x_train, y_train)

# Prediction
y_pred = BR.predict(x_test)
```

### **Evaluation**

```python
from sklearn.metrics import mean_squared_error
from sklearn import metrics
from sklearn.metrics import r2_score

MSE = mean_squared_error(y_test, y_pred)
R2 = metrics.r2_score(y_test, y_pred)

print("Mean Squared Error:", round(MSE, 4))
print("R2 score(TEST):", round(R2, 4))
```
