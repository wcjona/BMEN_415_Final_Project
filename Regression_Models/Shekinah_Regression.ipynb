```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error

df = pd.read_csv("/content/Volumetric_features.csv")
df.head()
```

## Preprocesing

```python
# Cleaning
df = df.drop(columns=["S.No", "dataset", "Left-WM-hypointensities", "Right-WM-hypointensities", "Left-non-WM-hypointensities", "Right-non-WM-hypointensities"])
df = df.drop_duplicates()
df = df.dropna()
df.head()
```

```python
# Split
y = df["Age"]
x = df.loc[:,df.columns!="Age"]

# Normalizing
x = (x-x.mean())/x.std()

x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=142)
```

# Regression Algorithms




 **Ordinary least squares Linear Regression**

```python
from sklearn.linear_model import LinearRegression

# Fit Regression Model
linearReg = LinearRegression().fit(x_train, y_train)

# Test Regression Model
y_pred = linearReg.predict(x_test)

print("R2: ", linearReg.score(x_test, y_test))
print("MSE: ", mean_squared_error(y_test, y_pred))
```

**Logistic Regression** 



```python
from sklearn.linear_model import OrthogonalMatchingPursuit

# Fit Model
omp = OrthogonalMatchingPursuit()
omp.fit(x_train, y_train)

# Test Model
y_pred = omp.predict(x_test)

print("R2: ", omp.score(x_test, y_test))
print("MSE: ", mean_squared_error(y_test, y_pred))
```

**Multilayer Perceptron Regressor**


```python
from sklearn.neural_network import MLPRegressor

# Fit Neural Net
mlp = MLPRegressor(max_iter=800).fit(x_train, y_train)

# Test Neural Net
y_pred = mlp.predict(x_test)
print("R2: ", mlp.score(x_test, y_test))
print("MSE: ", mean_squared_error(y_test, y_pred))
```
