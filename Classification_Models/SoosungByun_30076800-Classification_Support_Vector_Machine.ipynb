### **Data Curation**

```python
import pandas as pd

df = pd.read_csv("/content/Maternal Health Risk Data Set.csv")
df.head()
```

```python
df = df.drop_duplicates()
df = df.dropna()
df.head()
```

```python
df = df.replace({"low risk": 0, "mid risk":1, "high risk":2})
df.head()
```

```python
from sklearn.model_selection import train_test_split
y = df["RiskLevel"]
x = df.loc[:,df.columns!="RiskLevel"]

x.head()
```

```python
mean_std_x=(x-x.mean())/x.std()
mean_std_x.head()
```

```python
# Split training and test data
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=142)
```

### **Support Vector Machine Classification**

```python
# Importing necessary library
from sklearn.svm import SVC

# SVM classifier
SVM = SVC(kernel='linear')

# Training
SVM.fit(x_train, y_train)

# Prediction
y_pred = SVM.predict(x_test)
```

### **Evaluation**

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix

accuracy = accuracy_score(y_test, y_pred)
cm = confusion_matrix(y_test, y_pred)

print("Accuracy:", round(accuracy*100,4))
print("Confusion Matrix:\n",cm)

# For confusion matrix, the rows correspond to y_test and the columns correspond to y_pred
```
